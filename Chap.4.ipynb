```python
import matplotlib.pyplot as plt

#   /10 
x= [x/10 for x in range(20)]
y= [(x/10)**2 for x in range(20)]
z= [(x/10)**3 for x in range(20)]
p= [2**(x/10) for x in range(20)]

#x= [x for x in range(20)]
#y= [x**2 for x in range(20)]
#z= [x**3 for x in range(20)]
#p= [2**x for x in range(20)]


plt.plot(x, x, label= 'linear')
plt.plot(x, y, label= 'quadratic')
plt.plot(x, z, label= 'qubic')
plt.plot(x, p, label= 'power')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('My Plot')
plt.legend()
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

x=[]
y=[]
z=[]

for angle in range(360):
    x.append(angle)
    y.append(math.sin(math.radians(angle)))
    z.append(math.cos(math.radians(angle)))
    
plt.plot(x,y, label ='sin')
plt.plot(x,z, 'r-', label ='cos')

plt.title("SINE & COSINE WAVE")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x= np.linspace(0, (2*np.pi)*6)
y=[]
z=[]

for radian in x:
    y.append(radian*math.sin(radian))
    z.append(20*math.cos(radian))
    
plt.plot(x,y,'r-')
plt.plot(x,z, 'b--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 3, figsize= (15, 5))

x1= [77, 78, 85, 83, 73, 77, 73, 80]
y1= [25, 28, 29, 30, 21, 22, 17, 35]

x2= [75, 77, 86, 86, 79, 83, 83, 88]
y2= [56, 57, 50, 53, 60, 53, 49, 61]

x3= [34, 38, 38, 41, 30, 37, 41, 35]
y3= [22, 25, 19, 30, 21, 24, 28, 18]

ax[0].scatter(x1, y1, c='red')
ax[1].scatter(x2, y2, c='blue', marker= 's')
ax[2].scatter(x3, y3, c='green', marker= '^')

ax[0].set_title('Dachshund size')
ax[1].set_title('Samonyed size')
ax[2].set_title('Maltese size')

ax[0].set_xlabel('Length')
ax[0].set_ylabel('Height')
ax[1].set_xlabel('Length')
ax[1].set_ylabel('Height')
ax[2].set_xlabel('Length')
ax[2].set_ylabel('Height')

plt.show()
```
