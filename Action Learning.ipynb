```python
import matplotlib as mpl
import matplotlib.pylab as plt

#그래프 그림 사이즈
plt.rcParams.update({'figure.figsize': (10, 7), 'figure.dpi': 120})

#x는 개최연도,y는 관중 수
def plot_df(df, x, y, title="", xlabel='', ylabel='', dfi=100):
    plt.plot(x, y , marker="o", color='black')#그래프 그림, marker은 동그라미, 색은 검은색
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
    
plot_df(df, x=df.Year, y=df.Attendance, title='')#df에서 x는 Year, Y는 Attendance를 이용해서 그래프 구현
```
