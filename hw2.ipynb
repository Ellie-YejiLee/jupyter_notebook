```python
import pandas as pd #데이터프레임 라이브러리 import
from bs4 import BeautifulSoup as bs #뉴스 크롤링 라이브러리 import, 이는 bs로 정의
import requests #크롤링 라이브러리 impoort
import re #정규표현 라이브러리 import

key_word= input('검색 키워드를 입력하세요 : ') #검색 키워드 입력
count= int(input('총 필요한 뉴스기사 수를 입력해주세요(숫자만 입력) : ')) #필요한 뉴스기사 수 입력

html = requests.get('https://search.naver.com/search.naver?where=news&sm=tab_jum&query={}'.format(key_word))
#입력한 키워드를 {}에 넣은 URL을 requests의 get을 이용하여 html 코드로 받음
soup = bs(html.text, 'html.parser') #html.text를 bs(BeautifulSoup)함수로 파싱

news_dict= {} #뉴스기사 정보를 저장할 딕셔너리
idx = 0 #현재 뉴스기사 수

print()#줄 띄움
print('크롤링 중...')

while idx < count: #현재 뉴스기사 수가 필요한 뉴스기사 수보다 작을 경우 실행
    table = soup.find('ul',{'class' : 'list_news'})
    #ul태그, 해당 키워드에 대한 리스트로 정리된 뉴스
    li_list = table.find_all('li', {'id': re.compile('sp_nws.*')})
    #li태그, 해당 키워드에 대한 뉴스를 각 뉴스기사로
    area_list = [li.find('div', {'class' : 'news_area'}) for li in li_list]
    #div태그, 각 뉴스기사의 뉴스출처(신문사),뉴스제목, 본문에 대한 코드
    a_list = [area.find('a', {'class' : 'news_tit'}) for area in area_list]
    #a태그, 각 뉴스기사의 제목(title), 주소(url)에 대한 코드
    
    #a_list에 저장된 각 항목 n에 대해
    for n in a_list: 
        #필요한 뉴스기사 수가 현재 저장된 기사 수보다 클 경우(필요한 뉴스기사 수와 현재 저장된 뉴스기사 수가 같으면 반복 끝)
        if count-idx>0: 
            #news_dict에 현재 기사의 번호(수) 저장, 번호에 따른 title과 url 저장
            news_dict[idx] = {'title' : n.get('title'), #n에서 title에 해당되는 값
                          'url' : n.get('href')} #n에서 href에 해당되는 값
            idx += 1 #idx값 1 증가
        
print('크롤링 완료')
print('데이터프레임 변환')
news = pd.DataFrame(news_dict).T 
#저장한 뉴스기사를 데이터프레임으로 변환, .T로 행&열 변경
news #데이터프레임으로 저장된 정보 출력
```

```python
import pandas as pd #판다스
import seaborn as sns #시본

#시본에서 타이타닉 파일 호출
titanic = pd.read_csv('titanic.csv')

#성별 인원수
print('성별 인원수\n', titanic['Sex'].value_counts()) 
print()
#나이별 인원수
print('나이별 순서에 따른 인원수\n', titanic['Age'].value_counts().sort_index())
print()
#선실별 인원수
print('선실별 순서에 따른 인원수\n', titanic['Pclass'].value_counts().sort_index(ascending=False))
print()
#사망/생존 인원수
print('사망/생존 인원수(사망:0, 생존:1)\n', titanic['Survived'].value_counts())
```

```python
p1= titanic[titanic['Pclass'].isin([1])] #타이타닉 호 1등실 선실 승객
p2= p1[p1['Sex'].isin(['female'])] #1등실 선실 승객 중 여성
print('여성의 평균 나이는 ', p2.mean()['Age'] ,'이다') #여성의 나이 평균
```
