```python
import csv

f= open('C:/data/weather.csv')
data= csv.reader(f)
header= next(data)

max_wind=0.0

for row in data:
    if row[3]== '':
        wind = 0
    else:
        wind = float(row[3])
    if max_wind < wind :
        max_wind = wind
        
print(' 10    ', max_wind, 'm/s')
f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f= open('C:/data/weather.csv')
data= csv.reader(f)
header= next(data)

monthly_wind = [0 for x in range(12)]
days_counted = [0 for x in range(12)]

for row in data:
    month= int(row[0][5:7])
    if row[3] != '':
        wind = float(row[3])
        monthly_wind[month-1] += wind
        days_counted[month-1] += 1
        
for i in range(12):
    monthly_wind[i] /= days_counted[i]
    
plt.plot(monthly_wind, 'blue')
plt.show()

f.close()
```

```python
import pandas as pd

df1= pd.DataFrame({
    'name' : ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'horse power': [130, 250, 190, 300, 210, 220, 170],
    'weight': [1.9, 2.6, 2.2, 2.9, 2.4, 2.3, 2.2],
    'efficiency':[16.3, 10.2, 11.1, 7.1, 12.1, 13.2, 14.2] })
df1['com'] = 'P'

df2= pd.DataFrame({
    'name' : ['A', 'B', 'C', 'D'],
    'horse power': [120, 220, 120, 200],
    'weight': [1.9, 2.1, 1.5, 2.9],
    'efficiency':[18.3, 19.2, 21.1, 17.3] })
df2['com'] = 'Q'

df3= pd.concat([df1, df2])
df3= df3.set_index('name')
print(df3)
```

```python
df3['hpxmile']=df3['horse power']*df3['efficiency']
print(df3)
print(df3.groupby('com').mean()['hpxmile'])
```
